### Q.10. What is the maximum duration of a Movie/Show on Netflix ?

```python
sea_df=df[df.Duration.str.contains("Seasons")]
sea_df.head(10)
```

```python
sea_df["Number_of_Seasons"]=sea_df.Duration.str.replace("Seasons","")
sea_df["Number_of_Seasons"]=sea_df["Number_of_Seasons"].astype("int64")
sea_df.sort_values("Number_of_Seasons",ascending=False).head(6)
```

```python
min_df=df[df.Duration.str.contains("min")].reset_index()
min_df.head(10)
```

```python
min_df["minute_value"]=min_df["Duration"].str.replace("min","")
min_df["minute_value"]=min_df.minute_value.astype("int64")
min_df.sort_values("minute_value",ascending=False).head(6)
```
